```
paper_thickness = 0.00008
folding_num = 43
```

```
def folded_paper_calculator(paper_thickness, folding_num):
  result = paper_thickness * 2**folding_num
  return result
```

```
print(folded_paper_calculator(paper_thickness, folding_num))
```

## 問題１

```
def q1():
  """
  紙を1回折った時の厚さを計算するコード
  """
  THICKNESS = 0.00008
  folded_thickness = THICKNESS*2**43
  print("厚さ： {}メートル".format(folded_thickness))
```

```
q1()
```

## 問題２

```
# メートルをキロメートルに変換して小数点以下2桁で表示する
print("厚さ： {:.2f}万キロメートル".format(folded_thickness/1000/10000))
```

## 問題３

```
def q3():  
  x = 2
  for i in range(42):
    x *= 2
  folded_thickness = THICKNESS*x
  print("厚さ： {:.2f}万キロメートル".format(folded_thickness/1000/10000))
```

```
q3()
```

## 問題４

```
import time

start = time.time()
q1()
elapsed_time = time.time() - start
print("time : {}[s]".format(elapsed_time))
```

```
start = time.time()
q3()
elapsed_time = time.time() - start
print("time : {}[s]".format(elapsed_time))
```

## 問題５

```
def q5():  
  x = 1
  xlist = []
  xlist.append(THICKNESS)
  for i in range(43):
    x *= 2
    y = THICKNESS*x
    xlist.append(y)
  return xlist
```

```
xlist = q5()
print(len(xlist))
```

## 問題６

```
"""
グラフを表示する。タイトルと軸ラベル名付き。
"""
import matplotlib.pyplot as plt
%matplotlib inline
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(xlist) # 「リスト名」のところにリストの変数名を入れる
plt.show()
```
