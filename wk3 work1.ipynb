## 【問題7】特徴量間の関係の確認
特徴量間の関係を見ていきます。次の4つを行ってください。


- 縦軸と横軸に特徴量を１つずつ取り、種類毎に色分けした散布図を作る
- 全ての組み合わせの散布図を一度に表示させる、散布図行列を作る
- ４つの特徴量に関する相関係数行列を作る
- 相関係数行列のヒートマップを作る

相関係数行列はDataFrameのcorr()メソッドを使うことで求められます。

```python
fig5 = plt.figure()
ax5 = fig5.add_subplot(1, 1, 1)

## sepal_length-sepal_width
plt.scatter(df.iloc[0:51, :1], df.iloc[0:51, 1:2], label='setosa', color='red')
plt.scatter(df.iloc[51:101, :1], df.iloc[51:101, 1:2], label='versicolor', color='navy')
plt.scatter(df.iloc[101:151, :1], df.iloc[101:151, 1:2], label='virginica', color='green')

ax5.set_xlabel('sepal length [cm]')
ax5.set_ylabel('sepal width [cm]')

plt.show()
```

```python
fig6 = plt.figure()
sns.pairplot(df, hue='Species')
plt.show()
```

```python
corr_mat = df.corr(method='pearson')
corr_mat
```

```python
fig7 = plt.figure()
ax7 = fig7.add_subplot(1, 1, 1)

sns.heatmap(corr_mat,
            vmin=-1.0,
            vmax=1.0,
            center=0,
            annot=True, # True:格子の中に値を表示
            fmt='.1f',
            xticklabels=corr_mat.columns.values,
            yticklabels=corr_mat.columns.values
           )
plt.show()
```

相関係数行列から、がくの大きさと花びらの大きさ、花びらの幅には強い相関があることがわかる。がくの大きさと花びらの大きさの相関図から、３種類のアヤメの特徴は次のように言えるだろう。
- setosa：がく及び花びらの大きさが最も小さい傾向が強く、花びらの大きさは2cmを超えることはほとんどない。
- versicolor：がくの大きさが5~7cm程度、花びらの大きさが2~5.5cm程度。花びらの大きさは6cmを超えることはほとんどない。
- virginica：他のアヤメと比較し、がく、花びらは大きい傾向があり、最も大きさに個体差がある種類。がくは5~8cm程度。花びらは4~7.5cm程度である。
