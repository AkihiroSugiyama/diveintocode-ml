## 問題１　100日目のコメの数
100日目には合計で米粒をいくつもらっているのかを計算するコードを作成し、答えを出力してください。また、1日目から100日目までの「その日にもらう米の数」と「その日までに累計で貰う米の数」の推移を折れ線グラフで表現してください。（縦軸が米の数、横軸が日数）

```
import matplotlib.pyplot as plt
```

```
DEAFULT_RICE_NUM = 1
DEFAULT_COEFFICIENT = 2

rice_list = []
total_rice_list = []

for i in range(100):
  rice_num = DEFAULT_COEFFICIENT ** i
  rice_list.append(rice_num)
  total_rice = sum(rice_list)
  total_rice_list.append(total_rice)

print(rice_list)
print(len(rice_list))
print(total_rice_list)
print(len(total_rice_list))
```

```
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(rice_list) # 「リスト名」のところにリストの変数名を入れる
plt.show()
```

```
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(total_rice_list) # 「リスト名」のところにリストの変数名を入れる
plt.show()
```

## 問題２
この逸話は30日だったり81日だったりと日数が曖昧です。どの話にも対応できるように日数を引数とした 関数 を作成しましょう。関数を作成した上で、コースを受講する日数の間でいくつ米粒をもらえるかを答えてください。前の問題同様に折れ線グラフも出力してください。


以下の雛形には関数としての枠組みだけではなく、引数や返り値についてのコメントも記載してあります。参考にして、自分なりにも分かりやすくコメントを書き換えてください。

```
def compute_sorori_shinzaemon(day=100):
  rice_list = []
  total_rice_list = []

  for i in range(day):
    rice_num = DEFAULT_COEFFICIENT ** i
    rice_list.append(rice_num)
    total_rice = sum(rice_list)
    total_rice_list.append(total_rice) 
  return rice_list, total_rice_list

rice_list10, total_rice_list10 = compute_sorori_shinzaemon(day=10)
print(rice_list10)
print(total_rice_list10)
print(len(rice_list10))
print(len(total_rice_list10))
```

```
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(rice_list10) # 「リスト名」のところにリストの変数名を入れる
plt.show()
```

```
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(total_rice_list10) # 「リスト名」のところにリストの変数名を入れる
plt.show()
```

 ## 問題３　何人が何日生活できる米か
この米があれば何人が何日生活できるか計算する関数を作成してください。コースを受講する期間に貰える米粒で、このクラスにいる人数であれば、何日間生活できるかを出力してみましょう。


関数の最低限の入力は 米粒の数(int) と 人数(int) 、出力は 生活できる日数(int) です。関数名やコメントなどは分かりやすく考えて作成してください。


ここでは1日に必要な米粒の数は提示しません。自分なりに調べて、計算式に落とし込んでください。どう調べ、何を根拠にしたかもコメントを残しておいてください。

成人男性に必要な1日のコメの量：150g
[Reference](http://tokuteikenshin-hokensidou.jp/news/2019/008243.php#:~:text=1%E6%97%A5%E3%81%82%E3%81%9F%E3%82%8A150g(%E3%81%8A%E3%82%88%E3%81%9D,%E5%BA%A6%E3%81%8C%E4%BD%8E%E4%B8%8B%E3%81%99%E3%82%8B%E3%81%A8%E3%81%84%E3%81%86%E3%80%82&text=%E7%B1%B3%E3%81%AE%E6%91%82%E5%8F%96%E9%87%8F%E3%81%8C,%E3%81%A7%E6%98%8E%E3%82%89%E3%81%8B%E3%81%AB%E3%81%AA%E3%81%A3%E3%81%9F%E3%80%82)

米粒の重さ：0.02g
[Reference](https://www.komenet.jp/faq/ip17.html)

```
ADULT_MEN_RICE_AMOUNT_PER_DAY = 150 ## unit : g
RICE_GRAIN_WEIGHT = 0.02 ## unit : g
ADULT_MEN_RICE_GRAIN_PER_DAY = ADULT_MEN_RICE_AMOUNT_PER_DAY / RICE_GRAIN_WEIGHT

def compare(day, num_people):
  rice_list_x, total_rice_list_x = compute_sorori_shinzaemon(day)
  result = total_rice_list_x[-1] / num_people / ADULT_MEN_RICE_GRAIN_PER_DAY
  result = int(result)
  return day, num_people, result
```

```
day, num_people, result = compare(120, 15)
```

```
print("{}日間、{}人の場合{}日生存可能です".format(day, num_people, result))
```
